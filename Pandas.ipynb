***Merge and Join***

The pd.merge() function implements a number of types of joins: the
* one-to-one,
* many-to-one, and
* many-to-many joins.
All three types of joins are accessed by
identical call --> pd.merge()

###One-to-one joins

```python
df1 = pd.DataFrame({
    'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']
    })
df2 = pd.DataFrame({
    'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
    'hire_date': [2004, 2008, 2012, 2014]})
print(df1)
print(df2)
```

```python
df3 = pd.merge(df1, df2)
df3
```

```python
df4 = pd.concat([df1, df2])
df4
```

###*Many-to-one joins*

```python
df5 = pd.DataFrame({
    'group': ['Accounting', 'Accounting',
    'Engineering', 'Engineering',
    'HR', 'HR'],'skills': ['math', 'spreadsheets', 'coding', 'linux',
     'spreadsheets', 'organization']})
print(pd.merge(df1, df5))
```

***Specification of the Merge Key***

```python
print(pd.merge(df1, df2, on='employee'))
```

```python
df3 = pd.DataFrame({
    'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
    'salary': [70000, 80000, 120000, 90000]
    })
print(pd.merge(df1, df3, left_on="employee", right_on="name"))
```

```python
pd.merge(df1, df3, left_on="employee", right_on="name").drop('name', axis=1)
```

#**High-Performance Pandas: eval() and query()**

```python
df1=pd.DataFrame(np.random.randn(5, 4))
df2=pd.DataFrame(np.random.randn(5, 4))
df3=pd.DataFrame(np.random.randn(5, 4))
df4=pd.DataFrame(np.random.randn(5, 4))
df5=pd.DataFrame(np.random.randn(5, 4))
```

```python
result1 = -df1 * df2 / (df3 + df4) - df5
result2 = pd.eval('-df1 * df2 / (df3 + df4) - df5')
np.allclose(result1, result2)
```

```python
rng= np.random.RandomState(42)
df1,df2 = (pd.DataFrame(rng.randint(0,1000,(100,3)))for i in range(2))
r1=df1*df2
r2=pd.eval('df1*df2')
np.allclose(r1,r2)
```

```python
df = pd.DataFrame(rng.rand(1000, 3), columns=['A', 'B', 'C'])
df.head()
result1 = df[(df.A < 0.5) & (df.B < 0.5)]
result2 = df.query('A < 0.5 and B < 0.5')
np.allclose(result1, result2)
```
