```python
# my grade details
import pandas as pd
import numpy as np
SLNo=[1,2,3,4,5,6,7,8,9]
CGPA=pd.DataFrame({'Exam_Code':['22MTCO1','22PYC01','22CSCO1','22EGCO1','22PYCO3','22EGCO2','22CSCO2','22MEC01','22MEC38']
                   ,'Subject':['MATHS','PHYSICS','PPS','ENGLISH','PHYSICS LAB','ENGLISH LAB','PPS LAB','CAD & D','D&F'],
                   'FINAL_GRADE':['S','A','S','A','S','S','A','S','S'],
                   'Credits':[4,3,3,2,1.5,1,1.5,2.5,1.5]},index=SLNo)
Sno=CGPA.index.name
CGPA.index.name='SLNO'
print(CGPA)
print(CGPA.head())
print("\n")
print(CGPA.tail(4))
print()
print(CGPA.index)
print(CGPA.columns)
print(CGPA.to_numpy())
print(CGPA.describe())
print(CGPA.sort_values(by="Credits"))
print(CGPA.sort_index(axis=1,ascending=False))
print(CGPA.Credits)
print(CGPA[0:3])
print()
print(CGPA[1:4])
print(CGPA.loc[SLNo[0]])
print(CGPA.loc[:,["Exam_Code","FINAL_GRADE"]])
print(CGPA.loc["1":"2",["Subject","Credits"]])
print("physics grade is ",CGPA.loc[SLNo[1],"FINAL_GRADE"])
print("physics code is ",CGPA.at[SLNo[1],"Exam_Code"])
print(CGPA.iloc[3:5,0:2])
print(CGPA.iloc[[1,2,4],[0,2]])
print(CGPA.iloc[1:3,:])
print(CGPA.iloc[1,1])
print(CGPA.iat[2,2])
print(CGPA[CGPA["Credits"]<0])
CGPA_SEM1=CGPA.copy()
print(CGPA_SEM1)
CGPA_SEM1["status"]=["passed","passed","passed","passed","passed","passed","passed","passed","passed"]
print(CGPA_SEM1)
print(CGPA_SEM1[CGPA_SEM1['FINAL_GRADE'].isin(["S"])])
store=(10,9,10,9,10,10,9,10,10)
CGPA_SEM1["points"]=store
CGPA_SEM1
CGPA_SEM1.at[SLNo[0],"Exam_Code"]='22MTC02'
CGPA_SEM1
CGPA_SEM1.iat[0,1]="MATHEMATICS"
CGPA_SEM1
CGPA_SEM1.loc[:,"status"]=np.array(["PASSED"]*len(CGPA_SEM1))
CGPA_SEM1
CGPA_SEM1.mean()
CGPA_SEM1
```

```python
 import pandas as pd
Slno=[1,2,3,4]
phone=pd.DataFrame({'Mobile_number':[9876543210,9873216540,9874562130,9632587410],
                    'Name':['arun','tej','ram','ntr'],},index=Slno)
#tenth1=pd.Series(['yes'])
#inter=pd.Series(['yes'])
#phone["tenth"]=tenth1
phone.loc[Slno[0]:Slno[1],'Tenth_list']=['yes','no']
phone.loc[Slno[1]:Slno[3],'Inter_list']=['no','yes','yes']
print(phone)
print()
print(phone.dropna(how="any"))
print(phone.fillna(value='no'))
print()
print(phone)
print()
print(pd.isna(phone))
```
